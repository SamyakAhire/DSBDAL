```python
x = df.drop(columns=['species'],axis=1)
y = df['species']
y
```

```python
scaler = sk.preprocessing.MinMaxScaler()
x_scale = scaler.fit_transform(x)

x_scale
```

```python
train_test_split = sk.model_selection.train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scale,y,test_size=0.9,random_state=75)
```

```python
#from sklearn.naive_bayes import GaussianNB

gnb = sk.naive_bayes.GaussianNB()
gnb.fit(x_train,y_train)
y_predict = gnb.predict(x_test)
print(y_predict)

print(y_test)
```

```python
mx = ['setosa','virginica','versicolor']
#matrix = sk.metrics.confusion_matrix
# matrix(y_test,y_predict,labels=mx)
sk.metrics.confusion_matrix(y_test,y_predict,labels=mx)
```

```python
print(sk.metrics.classification_report(y_test,y_predict,labels=mx))
```
