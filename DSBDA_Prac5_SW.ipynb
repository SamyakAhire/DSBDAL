```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')

dataset.head()
```

```python
dataset.describe()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
sns.scatterplot(data=dataset,x='age',y='fare')
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
plt.title('Distribution of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
plt.title('Distribution of Fare by Passenger Class and Survival Status')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
sns.boxplot(x='pclass', y='fare', data=dataset, hue='survived')
```

```python
sns.violinplot(x='sex',y='age',data=dataset, hue='sex')
```

```python
sns.violinplot(x='sex',y='age',hue= 'survived',data=dataset)
```

```python
sns.violinplot(x='sex',y='age',hue= 'survived',split = True,data=dataset)
```

```python
sns.stripplot(x='sex',y='age',data=dataset, hue='sex')
```
