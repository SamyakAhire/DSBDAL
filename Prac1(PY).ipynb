```python
nc = ["Math_Score","Physics_Score","English_Score","Attendance"]

for c in nc:
    df[c] = df[c].fillna(df[c].mean())

df
```

```python
numeric_columns = ['Math_Score', 'Physics_Score', 'English_Score', 'Attendance']
for col in numeric_columns:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
numeric_columns = ['Math_Score','Physics_Score','English_Score', 'Attendance']
for col in numeric_columns:
    Q1 = df[col].quantile(0.28)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), df[col].median(), df[col])
df
```
