```python
def information(spd,spn):
    n_col = ['sepal_length','sepal_width','petal_length','petal_width']
    print(f'------------------------------{spn}-----------------------')
    print(f'mean: \n{spd[n_col].mean()}')
    print('-----------------------------------------------------------')
    print(f'median: \n{spd[n_col].median()}')
    print('-----------------------------------------------------------')
    print(f'25%: \n{spd[n_col].quantile(0.25)}')
    print('-----------------------------------------------------------')
    print(f'75%: \n{spd[n_col].quantile(0.75)}')
    print('-----------------------------------------------------------')
    print(f'std: \n{spd[n_col].std()}')
    print('-----------------------------------------------------------')
```

```python
information(setosa,'setosa')
```

```python
information(versicolor,'versicolor')
```

```python
information(virginica,'virginica')
```
